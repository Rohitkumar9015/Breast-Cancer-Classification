```python
# create a data frame
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df["target"] = cancer.target
```

```python
# Map 0/1 to string labels
df['target_label'] = df['target'].map({0: 'malignant', 1: 'benign'})
df.head()
```

```python
sns.pairplot(
    df,
    vars=["mean radius", "mean texture", "mean perimeter", "mean area", "mean smoothness"],
    hue="target_label",          # <- use the new column here
    palette={'malignant':'r',    # optional: give each class its own color
             'benign':'g'}
)
```

```python
sns.countplot(df["target_label"])
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target_label', data= df)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True) #, annote = True for no. inside cubes
```

```python
# first we divide data into input(x) and output(y)
x = df.iloc[:, :-2]
#x = df.drop(["target", "target_label"], axis=1)
x
```

```python
y = df["target"]
y
```

```python
# now we divide our data in tarining and testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state= 1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#model training

from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

model = SVC()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test) # model prediction
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict, target_names=['malignant', 'benign']))
```

```python
accuracy = accuracy_score(y_test, y_predict)*100
precision = precision_score(y_test, y_predict)*100
recall = recall_score(y_test, y_predict)*100
f1 = f1_score(y_test, y_predict)*100
print(f" Accuracy: {accuracy:.2f}\n Precision:  {precision:.2f} \n Recall :{recall:.2f} \n F1: {f1:.2f}")
```

```python
# STANDARDIZATION

from sklearn.preprocessing import StandardScaler

# Assuming your features are in X (excluding target column)
scaler = StandardScaler()

# If you have train-test split:
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# model training
model = SVC()
model.fit(X_train_scaled, y_train)

# model prediction
y_predict = model.predict(X_test_scaled) # model prediction

# evaluation of model
accuracy = accuracy_score(y_test, y_predict)*100
precision = precision_score(y_test, y_predict)*100
recall = recall_score(y_test, y_predict)*100
f1 = f1_score(y_test, y_predict)*100
print(f" Accuracy: {accuracy:.2f}\n Precision:  {precision:.2f} \n Recall :{recall:.2f} \n F1: {f1:.2f}")
```

```python
# SVM PARAMETERS

param_grid = {'C': [0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001], 'kernel':['rbf']}

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(SVC(), param_grid=param_grid, refit= True, verbose=4)
#param_grid : This tells GridSearchCV which hyperparameters to test.
#refit=True GridSearchCV finds the best combination of parameters, this will automatically retrain the SVC model using the entire training data with those best parameters.
#verbose=4 This just prints detailed progress while GridSearchCV is running (like which parameter it's trying, and the current fold in cross-validation).

grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_ # this will give best values
```

```python
grid_prediction = grid.predict(X_test_scaled) # doing prediction with the SVC with best parameter values
```

```python
# evaluation of model
accuracy = accuracy_score(y_test, grid_prediction)*100
precision = precision_score(y_test, grid_prediction)*100
recall = recall_score(y_test, grid_prediction)*100
f1 = f1_score(y_test, grid_prediction)*100
print(f" Accuracy: {accuracy:.2f}\n Precision:  {precision:.2f} \n Recall :{recall:.2f} \n F1: {f1:.2f}")
```

```python
# this checks the type 1 and type 2 error and no. of miss classifications done
cm = confusion_matrix(y_test, grid_prediction)
sns.heatmap(cm, annot=True)
```
